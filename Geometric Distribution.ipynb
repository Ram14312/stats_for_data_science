```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom
```

```python
# let us suppose someone is given the probability of success is 0.3
 
#for example let us suppose if i am trying to make a sales call on a dialy basis, if someone is going to ask us that
#how many calls we are supposed to do before we are going to make a very very first deal
# lets suppose probability of making a first deal is 0.3
#Q.Sales: How many calls before you close the first deal?
p = .3
# let us suppose we are creating a range using numpy function 
x = np.arange(1,20)
# 1,20 means 19 i.e, total number of trails is 19
# range means for suppose each call how much chance is present for successful deal
#let us take probability mass function
pmf = geom.pmf(x,p) # geometric funtion we are passing range and probability of success
```

```python
pmf

#in the output we can able to 19 results which is simply means that gives the probability of success in each call for closing a deal
#if someone is asking us a question that what is the probability that your first success is going to occur in fourth trail
#so it is going to be 10.29% or we can say 1029, so this is the chance of success
#what is the probability that you will be able to convert a call or you will able to close a deal on 19th call (i.e, last call)
#then the probability is 0.00048852 that we can able to successfully close the call
```

```python
#for plotting it out we can use plt
plt.bar(x,pmf)
#x axis represents number of trails and y axis represents probability of success for each trail
```
