```python
exercises = exercises.drop(columns=['Unnamed: 0'])
```

```python
#step 2: visualizing skewness, kurtosis and outliers

total_pulse = exercises['pulse']
```

```python
skew(total_pulse)
```

```python
#k_excess = kurtosis(x)        # normal = 0 as per programatical way 
#k_3based = k_excess + 3       # normal = 3 as per textbook definition
kurtosis(total_pulse)
```

```python
# IQR (inter quantile range to find out outliers)

q1 = total_pulse.quantile(.25)
```

```python
q1
```

```python
q2 = total_pulse.median()
```

```python
q3 = total_pulse.quantile(.75)
```

```python
q3
```

```python
iqr = q3-q1
```

```python
iqr
```

```python
lower_bound = q1 -1.5*iqr
```

```python
lower_bound
```

```python
upper_bound = q3 + 1.5*iqr
```

```python
upper_bound
```

```python
# visualizing with histogram

plt.figure(figsize =(10,5))
```

```python
# to build correlation matrix

ex_num = exercises[['pulse']]

ex_dummies = p.get_dummies(
    exercises[['diet','time','kind']],
    drop_first=True
)
```

```python
corr_df = p.concat([ex_num,ex_dummies], axis=1)
corr = corr_df.corr(numeric_only=True)
```
