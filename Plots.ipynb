```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew,mode
```

```python
#in this inbuilt-library we will have datasets like tips dataset

tips = sns.load_dataset('tips')
```

```python
tips
```

```python
type(tips)
```

```python
#here type seems to be DataFrame simply means that data is already available ina pandas data frame and pandas like operation
#so we will be able to perform in the easiest possible way
#first from this tips data frame we can extract the total_bill
total_bill = tips['total_bill']
```

```python
total_bill
```

```python
# to find average of the data
total_bill.mean()
```

```python
#to find median
total_bill.median()
```

```python
#to find mode
total_bill.mode()
```

```python
#to skewness in the data
skew(total_bill)
# here the value has positive skewness i.e, in bills there is some one paid many bills so thats why we got a positive skewness
```

```python
q1 = total_bill.quantile(.25)
```

```python
q1
```

```python
q2 = total_bill.median()
```

```python
q2
```

```python
q3 = total_bill.quantile(.75)
```

```python
q3
```

```python
iqr = q3-q1
```

```python
iqr
```

```python
lower_bound = q1 - 1.5*iqr
```

```python
lower_bound
```

```python
upper_bound = q3 +1.5*iqr
```

```python
upper_bound
```

```python
#histogram
#to draw a histogram
plt.figure(figsize=(10,5))
```

```python
sns.histplot(total_bill)
```

```python
#histogram is a basic plot it is going to help us to finding out  shape of the data
#i.e, when our dataset is normal or our data set is skewed 
# x -axis talk about range of distribution of data
# y- axis is going to talk about occurence of the data  i.e how many people were paid

# to understand this witha question like how many people have apid a bill between $30 to $33 

# if connect all of the tips we can able to know about shape of the normal distribution curve i.e, with the help of the graph 
#we can able to find out the skewness
```

```python
sns.histplot(total_bill, kde=True, color='red',bins=10)
#bins is going to change shape of plots
#kde=true is going to build the curve
#color property is going to add color to the plot
```

```python
#box plot
sns.boxplot(x=total_bill)
```

```python
# in the box plot we can able to see some circles which looks like a outlier
# it is used to to see IQR 
# we can able to find q1, q2, q3, iqr , lower_bound and upper_bound from the box plot
```

```python
sns.barplot(x=tips['day'],y= tips['tip'], color='yellow', estimator= np.mean)
# estimator we can able to find what was the average of tips in a day
```

```python
#pie-chart
# the smoker-distribution can be explained with pie-chart
#i.e to provide how many smokers and non-smokers
smoker_count = tips['smoker'].value_counts()
```

```python
smoker_count
```

```python
plt.pie(smoker_count)

#blue is non-smoker
#orange is smoker just we can able to understand by seeing inorder to viewer understand the things we can include label
```

```python
plt.pie(smoker_count, labels=smoker_count.index)
```

```python
#pair-plot
sns.pairplot(tips[['total_bill','tip','size']])
```

```python
# thhe pair plot of total_bill it is going to show us the scatter plot , histogram, the pairplot is going to 
#provide lot of understanding between column to column
#pair plot is going to provid ethe relation between total_bill to size, tip, total_bill 
#similarly it is going to plot for tip to size, tip, total_bill
```

```python
#heat map
#correlation between the graphs
corr = tips[['total_bill', 'tip', 'size']].corr()
sns.heatmap(corr, annot=True , cmap = 'coolwarm', fmt = '.2f', linewidths=0.5)
```
